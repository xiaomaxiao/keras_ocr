```python
import numpy as np  

lines=[]
with open('loss3.cvs','r') as f:
    lines = f.readlines()
    lines = np.array(lines)
    lines = lines[lines!='\n']
    lines = np.array([i[:-1].split(',') for i in lines])
    lines = lines.astype('float32')
    

    
```

```python
import matplotlib.pyplot as plt  

beg = 0
step=100
cls_loss = lines[:,0][beg:-1:step]
regr_loss = lines[:,1][beg:-1:step]

clip =10

cls_loss[cls_loss>clip]=clip
regr_loss[regr_loss>clip]=clip

x = range(0,len(cls_loss))  
plt.figure(figsize=(10,4)) #
plt.plot(x,cls_loss,"r-",linewidth=1)  
plt.plot(x,regr_loss,"b-",linewidth=1)  
plt.title("loss") #  
plt.show()  #  
```
